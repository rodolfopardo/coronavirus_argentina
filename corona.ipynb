```python
from pytrends.request import TrendReq
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.io.json import json_normalize

sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
#Keywords coronavirus
pytrend = TrendReq()
pytrend.build_payload(kw_list=['Coronavirus'], geo = 'AR', timeframe = 'now 1-d')
```

```python
pytrend = TrendReq()
pytrend.build_payload(kw_list=['Coronavirus', 'sintomas', 'contagio', 'fiebre'], geo = 'AR', timeframe = 'now 1-d')
interest_over_time_df = pytrend.interest_over_time()
```

```python
interest_over_time_df = pytrend.interest_over_time()
```

```python
print(interest_over_time_df.tail())
```

```python
interest_by_region_df = pytrend.interest_by_region()
```

```python
interest_by_region_df.sort_values(by = 'Coronavirus', ascending = True).plot.barh()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes= False)
dx = interest_over_time_df.plot.line(figsize = (12,9),title = "Coronavirus")
dx.set_xlabel('Hora en GMT 0', fontsize = 15)
dx.set_ylabel('Bsquedas',fontsize = 15)
dx.tick_params(axis='both', which='major', labelsize=13)
```
